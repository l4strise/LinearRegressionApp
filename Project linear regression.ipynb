```python
import tkinter as tk
from tkinter import messagebox
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
class LinearRegressionApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Linear Regression App")
        self.X_label = tk.Label(root, text="Independent variable (X):")
        self.X_label.pack()
        self.X_entry = tk.Entry(root, width=40)
        self.X_entry.pack()
        self.y_label = tk.Label(root, text="Dependent variable (y):")
        self.y_label.pack()
        self.y_entry = tk.Entry(root, width=40)
        self.y_entry.pack()

        self.run_button = tk.Button(root, text="Run Model", command=self.run_model)
        self.run_button.pack()

        self.coeff_label = tk.Label(root, text="Coefficients:")
        self.coeff_label.pack()
        self.coeff_text = tk.Text(root, width=40, height=2)
        self.coeff_text.pack()
        
    def run_model(self):
        X_input = self.X_entry.get()
        y_input = self.y_entry.get()
        X = np.array([float(x) for x in X_input.split(",")]).reshape(-1, 1)
        y = np.array([float(y) for y in y_input.split(",")])

        model = LinearRegression()
        model.fit(X, y)

        self.coeff_text.delete(1.0, tk.END)
        self.coeff_text.insert(tk.END, f"Intercept: {model.intercept_:.2f}\nSlope: {model.coef_[0]:.2f}")

        plot_window = tk.Toplevel(self.root)
        plot_window.title("Linear Regression Plot")
        canvas = tk.Canvas(plot_window, width=400, height=300)
        canvas.pack()
        fig, ax = plt.subplots(figsize=(4, 3))

        ax.scatter(X, y, label='Original data')
        y_pred = model.predict(X)
        ax.plot(X, y_pred, color='red', label='Regression line')

        ax.set_title('Linear Regression')
        ax.set_xlabel('Independent variable (X)')
        ax.set_ylabel('Dependent variable (y)')
        ax.legend()

        canvas.figure = fig
        fig.canvas.mpl_connect('close_event', lambda event: plot_window.destroy())
        fig.canvas.draw()
        canvas.create_window(0, 0, window=fig.canvas, anchor='nw')
```

```python
if __name__ == "__main__":
    root = tk.Tk()
    app = LinearRegressionApp(root)
    root.mainloop()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

X_input = input("Enter independent variable values (separated by commas): ")
y_input = input("Enter dependent variable values (separated by commas): ")

X = np.array([float(x) for x in X_input.split(",")]).reshape(-1, 1)
y = np.array([float(y) for y in y_input.split(",")])
model = LinearRegression()
model.fit(X, y)

print("Coefficients:")
print("Intercept:", model.intercept_)
print("Slope:", model.coef_)

y_pred = model.predict(X)

plt.scatter(X, y, label='Original data')

plt.plot(X, y_pred, color='red', label='Regression line')
plt.title('Linear Regression')
plt.xlabel('Independent variable (X)')
plt.ylabel('Dependent variable (y)')
plt.legend()
plt.show()
```
